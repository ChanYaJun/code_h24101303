```python
#########################################################
from sklearn.svm import SVC
#linear

svc_linear=SVC(kernel='linear')  
svc_linear.fit(tr_i,tr_y)


y_pred=svc_linear.predict(te_i)

print(svc_linear.score(tr_i,tr_y))
print(svc_linear.score(te_i,te_y))


#poly
svc_poly=SVC(kernel='poly')  
svc_poly.fit(tr_i,tr_y)


y_pred=svc_poly.predict(te_i)

print(svc_poly.score(tr_i,tr_y))
print(svc_poly.score(te_i,te_y))

#rbf
svc_rbf=SVC(kernel='rbf')  
svc_rbf.fit(tr_i,tr_y)


y_pred=svc_rbf.predict(te_i)

print(svc_rbf.score(tr_i,tr_y))
print(svc_rbf.score(te_i,te_y))
```

```python
param_grid = {'C':[1,2,3,4,6,8,10,12,14],'gamma':[0.05,0.5, 1, 2], 'kernel':['rbf']}

grid = GridSearchCV(SVC(),param_grid,refit = True, verbose=2)
grid.fit(tr_i,tr_y)

print(grid.best_estimator_)
#SVC(C=1, gamma=0.5)

print(grid.score(tr_i,tr_y))
print(grid.score(te_i,te_y))
###############################################################
```

```python
#Tree
##########################################################
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(tr_i, tr_y)
Y_pred = dtc.predict(te_i)


print(confusion_matrix(te_y, Y_pred))
print(dtc.score(tr_i,tr_y))
print(dtc.score(te_i,te_y))

```

```python
#grid + tree 

grid_params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [3, 5, 7, 10],
    'min_samples_split' : range(2, 10, 1),
    'min_samples_leaf' : range(2, 10, 1)
}
grid_search = GridSearchCV(dtc, grid_params, cv = 5, n_jobs = -1, verbose = 1)
grid_search.fit(tr_i, tr_y)
dtc = grid_search.best_estimator_
Y_pred = dtc.predict(te_i)  

print(confusion_matrix(te_y, Y_pred))

print(grid_search.score(tr_i,tr_y))
print(grid_search.score(te_i,te_y))
print(grid_search.best_estimator_)
#############################################################
```
